# Data types in Python
ประเภทข้อมูลใน Python

##Numbers
- `int` for integers
- `float` for real numbers. They can represent decimal points.
- `complex` for complex numbers(with real and imaginary parts,very scientific!)

To represent `int` values in the program,type *numbers* without decimal points.

````python
print(32)
````

`type()`can be used for checking the data types of any values.

````python
print(type(32))
````

To represent `float` values in the program, type *numbers with decimal points*.

````python
print(5.0)
````

To represent `complex` values in the program, use the format `a+bj`.

````python
print(5+3j)
````

## Strings
A string object is a sequence of chracters. In Python, the data type is `str`.
To represent `str` in the program,enclose text with quotes.

Quotes could be:
- Single quotes
- Dobule quotes
- Triple quotes

````python
print('Hello')
print("Hello")
print('''Hello''')
print(type('Hello'))
print(type("Hello"))
print(type('''Hello'''))
````

Since quotes are used for singnaling the starting/ending of a String object, how can we write a sentence
like *He said "Hello"* or *It's ok to say "Goodbye"*.

````python
# print("He said "Hello"") คอมพิวเตอร์จะสับสน ว่าเรา quotes ข้อความอันไหน ดังนั้นจึงต้องแก้
print('He said "Hello".') #โดยการใช้ quotes ที่แตกต่างกัน
print('''It's ok to say "Goodbye".''')
````

If switching quotes is NOT a choice(ไม่ต้องการเปลี่ยน quotes),we can *escape* the quote character with a backslash.

````python
print("He said \"Hello\"")
print("It's ok to say \"Goodbye\".")
#พิมพ์ \ เพื่อบอกว่าเป็น Double quotes.
````

## Unary Minus
A unary minus *inverses the polarity* of the numeric operand

````python
a = 2
print(-a)
b = -5
print(-b)
````

## Modulus
A modulus (`a % b` reads as "a modulo b") finds the *remainder* resulting from dividing the left operand (`a`) with the right operand (`b`).

````python
x = 25
print(x % 4) # Since 25 = 4*6 + 1
print(x % 5) # Since 25 = 5*5 + 0
print(x % 7) # Since 25 = 7*3 + 4
````

## Exponentiation
An exponentiation (`a ** b`)finds the value of the left operand(`a`)raised to the power of the value of the right operand(`b`).

````python
print(2 ** 3)
print(5 ** 4)
print(29 ** 0) # Since any numbers raised to the power of 0 is 1
print(2 ** -1) # Since it is 1/2
````

### Floor Division
A floor division (`//`)is a *division* that *rounds* the result down to the nearest *integer*(the next smaller integer).

````python
print(32 / 3)
print(32 // 3)
print(-32 / 3)
print(-32 // 3)
````

## Comparison Operators
A comparison operator *compares two values*. it returns `True` or `False`

````python
a = 20
print(a == 20) # is equal to ?
print(a != 20) # is NOT equal to ?
print(a > 20) # is grater than ?
print(a < 20) # is less than ?
print(a >= 20) # is greater than or equal to ?
print(a <= 20) # is less than or equal to ?
````

## Logical Operators
A logical operator performs a *boolean operation*(AND,OR,NOT)

````python
p = True
q = False
# and
print(p and p)
print(p and q)
print(q and q)
# or
print(p or p)
print(p or q)
print(q or q)
# not
print(not p)
print(not q)
````
